```python
import os
from dotenv import load_dotenv
import requests
```

```python
load_dotenv()
BASE = os.environ.get("BASE")
TEAM_ID = os.environ.get("TEAM_ID")
```

```python
endpoint = f"/entry/{TEAM_ID}/"
response = requests.get(BASE+endpoint)
print(response.status_code)
```
